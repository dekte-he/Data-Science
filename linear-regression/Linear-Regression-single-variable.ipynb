```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
import numpy as np
from sklearn.preprocessing import StandardScaler
import math
```

```python
df=pd.read_csv("headbrain.csv")
df.head()
```

```python
scaler=StandardScaler()
df=scaler.fit_transform(df)
df=pd.DataFrame(data=df,columns=['Gender','Age Range','Head Size(cm^3)','Brain Weight(grams)'])
df.head()
```

```python
x=df['Head Size(cm^3)']
y=df['Brain Weight(grams)']
```

```python
plt.scatter(x,y)
```

```python
mean_x=x.mean()
mean_y=y.mean()
```

```python
slope=((x-mean_x)*(y-mean_y)).sum()/((x-mean_x)*(x-mean_x)).sum()
intercept=mean_y-slope*mean_x
```

```python
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.plot(x, y, '.')
plt.plot(x, intercept+ slope * x, '-')
plt.legend()
plt.show()
```

```python
y_predicted=slope*x+intercept
slope,intercept
```

```python
r_square=((y_predicted-mean_y)**2).sum()/((y-mean_y)**2).sum()
r_square
```

```python
iteration=1000
learning_rate=.01
n=len(y)
slope=0
intercept=0
for i in range(iteration):
    y_predicted=slope*x+intercept
    diff1=-2*((y-y_predicted)*x).sum()/n
    diff2=-2*(y-y_predicted).sum()/n
    slope=slope-learning_rate*diff1
    intercept=intercept-learning_rate*diff2
    cost=((y-y_predicted)**2).sum()/n
    print("slope",slope,"  intercept",intercept,"  iteration",i,"  cost",cost)
    
```

```python
r_square=((y_predicted-mean_y)**2).sum()/((y-mean_y)**2).sum()
r_square
```

```python
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.plot(x, y, '.')
plt.plot(x, intercept+ slope * x, '-')
plt.legend()
plt.show()
```

```python
data=pd.read_csv("marks.csv")
data.head()
```

```python
x=data['math']
y=data['cs']
```

```python
plt.plot(x,y,'.')
plt.xlabel('math')
plt.ylabel('cs')
```

```python
iteration=1000
slope=0
intercept=0
learning_rate=.0002
cost_previous=0
for i in range(iteration):
    y_predicted=slope*x+intercept
    diff1=-(2/n)*sum((y-y_predicted)*x)
    diff2=-(2/n)*sum((y-y_predicted))
    slope=slope-learning_rate*diff1
    intercept=intercept-learning_rate*diff2
    cost=((y-y_predicted)**2).sum()/n
    if math.isclose(cost, cost_previous, rel_tol=1e-20):
        break
    cost_previous=cost
    print("slope",slope,"  intercept",intercept,"  iteration",i,"  cost",cost)
```

```python
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.plot(x, y, '.')
plt.plot(x, intercept+ slope * x, '-')
plt.legend()
plt.show()
```
